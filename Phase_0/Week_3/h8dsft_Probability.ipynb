```python
# Buat fungsi pelemparan koin sesuai dengan nilai input yang ingin dimasukkan.

def coinflip(input):
    cf_result = []
    
    for i in range(input):
        flip = random.randint (0, 1)
        if (flip == 0):
            flip_result = "Head"
        else:
            flip_result = "Tail"
        cf_result.append(flip_result)
        
    print(str(cf_result))
    print("Jumlah head:", str(cf_result.count("Head")))
    print("Jumlah tail:", str(cf_result.count("Tail")))
```
