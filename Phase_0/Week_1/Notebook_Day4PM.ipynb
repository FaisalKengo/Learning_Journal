# Visualization

```python
# Visualisasi plot
# Yang dipanggil pertama merupakan variabel x

nba[nba["year_id"] > 2010].groupby("year_id")["pts"].max().plot()
```

```python
# Latihan

# 1. Siapkan list kosong untuk tempat penampung

hasil = []

# 2. Siapkan argumen looping
# 3. Masukkan hasil looping ke dalam list kosong

for baris in range(len(nba)):
    selisih = nba.iloc[baris, 10] - nba.iloc[baris, 14]
    hasil.append(selisih)

# 4. Buat list tersebut menjadi kolom baru

nba["diff_loop"] = hasil
```

```python
nba.head()
```

```python
def wld(diff):
    if diff > 0:
        return "W"
    elif diff < 0:
        return "L"
    else:
        return "D"

keterangan = []
for idx in range(len(nba)):
    res = wld(nba.loc[idx, "pts_diff"])
    keterangan.append(res)

nba["status"] = keterangan
```
